```python
import cv2
from yolo import count_people, count_satellite, predict_and_detect, count_people
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/image_class.jpg")
result_img, count = count_people(img, 0.5)
plt.imshow(result_img[...,::-1])
plt.show()
```

```python
image = cv2.imread("data/port.png")
result_img, count = count_satellite(image, [2], conf =0.7)
plt.imshow(result_img[...,::-1])
plt.show()
```

```python
image = cv2.imread("data/singapore.png")
result_img, count = count_satellite(image, [1], conf =0.25)
print(f'{count} ships detected')
plt.imshow(result_img[...,::-1])
plt.show()
```

```python
image = cv2.imread("data/fields.png")
model = YOLO("yolo11x.pt")
result_img, _, count = predict_and_detect(model, image, classes=[], conf=0.2)
plt.imshow(result_img[...,::-1])
plt.show()
```
